## 8. Introducing a Biological Dataset
We will create a synthetic OD (optical density) growth curve. This models how bacterial cultures grow over time.

[Graphics idea: conceptual diagram of exponential bacterial growth]


````python
import numpy as np

time = np.array([0, 1, 2, 3, 4, 5, 6, 7])
od = np.array([0.05, 0.06, 0.08, 0.12, 0.20, 0.33, 0.52, 0.82])

time, od
````

### Inspecting the Data
We can ask basic questions about the dataset.

````python
print("Time points:", time)
print("OD values: ", od)
print("Length of dataset:", len(time))
````

## 9. Visualizing the Growth Curve
Data becomes meaningful when visualized. We'll use matplotlib for plotting.

[Graphics idea: high quality OD curve plot with labeled axes]


````python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,4))
plt.plot(time, od, marker='o', linestyle='-', color='navy')
plt.xlabel('Time (hours)')
plt.ylabel('Optical Density (OD)')
plt.title('Synthetic Bacterial Growth Curve')
plt.grid(True)
plt.show()
````

## 10. Adding Mathematical Transformations
Let's compute the fold increase in OD relative to the starting point.

````python
fold = od / od[0]
fold
````

Plotting the fold increase helps visualize early growth.


````python
plt.figure(figsize=(7,4))
plt.plot(time, fold, marker='s')
plt.xlabel('Time (hours)')
plt.ylabel('Fold Increase')
plt.title('Growth Fold Change Over Time')
plt.grid(True)
plt.show()
````

## 12. Comments Make Code Understandable
A comment begins with `#` and is ignored by Python. Use them to explain logic.


````python
# Compute the doubling time approximately
approx_doubling_time = time[np.argmax(od > 2*od[0])] if np.any(od > 2*od[0]) else None
approx_doubling_time
````
